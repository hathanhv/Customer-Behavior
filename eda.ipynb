```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import silhouette_score


df['transaction_date'] = pd.to_datetime(df['transaction_date'])
today = df["transaction_date"].max()

# 3. Feature Engineering: đặc trưng theo user
user_df = df.groupby("user_id").agg(
    num_transactions=('transaction_id', 'count'),
    #total_amount=('product_amount', 'sum'),
    #avg_amount=('product_amount', 'mean'),
    #avg_transaction_fee=('transaction_fee', 'mean'),
    #avg_cashback=('cashback', 'mean'),
    #total_net_value=('Net Value', 'sum'),
    total_loyalty_points=('loyalty_points', 'sum'),
    total_net_value=('Net Value', 'mean'),
    #unique_merchants=('merchant_id', pd.Series.nunique),
    #unique_categories=('product_category', pd.Series.nunique), 
    #active_days=('transaction_date', lambda x: (x.max() - x.min()).days + 1),
    days_since_last = ('transaction_date', lambda x: (today - x.max()).days),
    # weekend_txns = ('date_of_week', lambda x: x.isin(['Saturday', 'Sunday']).sum())
).reset_index()

# 4. Mã hóa một số biến nếu cần
# (Có thể bổ sung encode payment_method phổ biến nếu thấy hợp lý)

# 5. Chuẩn hóa dữ liệu
features = user_df.drop(columns=['user_id'])
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

inertia = []
silhouette_scores = []
k_range = range(2, 11)  # silhouette không định nghĩa cho k=1

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(scaled_features)
    
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(scaled_features, labels))

# Vẽ biểu đồ
fig, ax1 = plt.subplots(figsize=(10, 5))

# Elbow
ax1.set_title('Elbow & Silhouette Score')
ax1.plot(k_range, inertia, 'bo-', label='Inertia (Elbow)')
ax1.set_xlabel('Số cụm (k)')
ax1.set_ylabel('Inertia', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Silhouette (vẽ trên y2)
ax2 = ax1.twinx()
ax2.plot(k_range, silhouette_scores, 'ro-', label='Silhouette Score')
ax2.set_ylabel('Silhouette Score', color='r')
ax2.tick_params(axis='y', labelcolor='r')

fig.tight_layout()
plt.grid(True)
plt.show()
```

```python

# 6. Giảm chiều với PCA để trực quan
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)
user_df['pca1'] = pca_result[:, 0]
user_df['pca2'] = pca_result[:, 1]

# 7. KMeans clustering
k = 3  # Bạn có thể dùng elbow method để chọn số cụm tốt hơn
kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
user_df['cluster'] = kmeans.fit_predict(scaled_features)

# 8. Vẽ biểu đồ kết quả phân cụm
plt.figure(figsize=(10, 6))
sns.scatterplot(data=user_df, x='pca1', y='pca2', hue='cluster', palette='Set2')
plt.title("Customer Segments (PCA + KMeans)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster")
plt.grid(True)
plt.tight_layout()
plt.show()

# 9. Thống kê đặc trưng trung bình theo cụm
cluster_profile = user_df.groupby('cluster').mean(numeric_only=True)
print("📊 Đặc trưng trung bình theo cụm:")
print(cluster_profile)

# 10. (Tuỳ chọn) Xuất ra CSV để dùng Power BI
user_df.to_csv("customer_segments.csv", index=False)
```
